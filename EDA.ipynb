```python
import geopandas as gpd
import numpy as np
import pandas as pd
import plotly.express as px
```

# Life Expectancy At Birth

```python
df = pd.read_csv('source/lifeExpectancyAtBirth.csv')
df.sort_values('Period')
```

```python
# There are no values for 2020 and all the values for Canada seem to be present, I will drop this observatoin
df.drop(index=372,inplace=True)
```

```python
px.choropleth(data_frame=df,
              locations='Location',
              locationmode='country names',
              animation_frame='Period',
              category_orders={'frame': [2000, 2010, 2015, 2019]},
              color='First Tooltip',
              color_continuous_scale='plotly3',
              title=df['Indicator'][0],
              facet_row='Dim1',
              height=1500
             )
```

```python
px.bar(data_frame=df.sort_values('First Tooltip'),
       x='Location',
       y='First Tooltip',
       animation_frame='Period',
       category_orders={'frame': [2000, 2010, 2015, 2019]},
       title=df['Indicator'][0],
       facet_row='Dim1',
       height=1500
      )
```
